```python
with open('20.txt', 'r') as f:
    data=tuple(int(line) for line in f.read().splitlines())
    
values={}
for i in range(len(data)):
    values[i]=data[i]
values
valueslist=tuple(i for i in range(len(data)))
valueslist

#item:location
sequence={}
#location:item
sequencereverse={}

for i in range(len(valueslist)):
    sequence[valueslist[i]]=i+1
    sequencereverse[i+1]=valueslist[i]

def movesequence(what,where):
    was=sequence[what]
    sequence[what]=where
    items=set()
    items.add(what)
    if where>was:
        for i in range(was+1,where+1):
            itemati=sequencereverse[i]
            items.add(itemati)
            sequence[itemati]-=1
        for i in items:
            sequencereverse[sequence[i]]=i
    elif was>where:
        for i in range(where,was):
            itemati=sequencereverse[i]
            items.add(itemati)
            sequence[itemati]+=1
        for i in items:
            sequencereverse[sequence[i]]=i

#print(len(data))
for i in range(len(data)):
    #decoded=[]
    move=values[i]
    where=sequence[i]+move
    while where<=0:
        where=len(data)+where-1
    while where>len(data):
        where=1+where-len(data)
    #print(i,values[i],where)
    movesequence(i,where)
    #for j in range(1,len(data)+1):
    #    decoded.append(values[sequencereverse[j]])
    #print(decoded)

decoded=[]    
for j in range(1,len(data)+1):
    decoded.append(values[sequencereverse[j]])
#print(decoded)
summ=decoded[(decoded.index(0)+1000)%len(data)]+decoded[(decoded.index(0)+2000)%len(data)]+decoded[(decoded.index(0)+3000)%len(data)]
#print(decoded[(decoded.index(0)+1000)%len(data)],decoded[(decoded.index(0)+2000)%len(data)],decoded[(decoded.index(0)+3000)%len(data)])
print(summ)
```
