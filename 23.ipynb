```python
with open('23.txt', 'r') as f:
    data=tuple(line for line in f.read().splitlines())

    
elves={}
elvesset=set()
elfnum=1
for i in range(len(data)):
    for j in range(len(data[i])):
        if data[i][j]=="#":
            elves[elfnum]=(j,i)
            elfnum+=1
            elvesset.add((j,i))

elvesproposed={}
#if num = 0 we check north first,1 south,2 west,3 east
def proposedmaker(location , num):
    i,j = location[0],location[1]
    N, NW, NE = (i,j-1),(i-1,j-1),(i+1,j-1)
    S, SW, SE = (i,j+1),(i-1,j+1),(i+1,j+1)
    E, W = (i+1,j), (i-1,j)
    alll=(N,NW,NE,S,SE,SW,E,W)
    
    hasneighbour=False
    for i in alll:
        if i in elvesset:
            hasneighbour=True
            break
    if not hasneighbour: return location

    check = ((N, NW, NE) , (S, SW, SE) , (W, SW, NW) , (E, SE, NE))
    
    for i in range(num,num+4):
        if i>3:
            i=i-4
        for j in check[i]:
            empty=True
            if j in elvesset:
                empty=False
                break
        if empty:
            return check[i][0]
        
    return location
        
def printpicture(elvesset):
    x0,x1=0,10
    y0,y1=0,10
    for i in elvesset:
        if i[0]<x0:
            x0=i[0]
        elif i[0]>x1:
            x1=i[0]
        if i[1]<y0:
            y0=i[1]
        elif i[1]>y1:
            y1=i[1]
    #if x1>10:print("expanded")
    for i in range(y0,y1+1):
        a=""
        for j in range(x0,x1+1):
            if (j,i) in elvesset:
                a+='#'
            else:
                a+='.'
        print(a)
        
def empties(elvesset):
    first=True
    for i in elvesset:
        if first:
            x0,x1=i[0],i[0]
            y0,y1=i[1],i[1]
            first=False
            continue
        if i[0]<x0:
            x0=i[0]
        elif i[0]>x1:
            x1=i[0]
        if i[1]<y0:
            y0=i[1]
        elif i[1]>y1:
            y1=i[1]
    
    a=0
    for i in range(y0,y1+1): 
        for j in range(x0,x1+1):
            if (j,i) not in elvesset:
                a+=1
    return a

cycle=-1
#print(elves)
#print(elvesset)
for i in range(10):
    cycle+=1
    if cycle==4:cycle=0
    #print(cycle)
    #printpicture(elvesset)
    #print("")
    proposedset=set()
    bannedset=set()
    for j in range(1,len(elves)+1):
        elvesproposed[j] = proposedmaker(elves[j],cycle)
        #print(elvesproposed[j])
        if elvesproposed[j] not in proposedset: proposedset.add(elvesproposed[j])
        else: bannedset.add(elvesproposed[j])
    elvesset=set()
    for j in range(1,len(elves)+1):
        if elvesproposed[j] not in bannedset:
            elves[j]=elvesproposed[j]
        elvesset.add(elves[j])
    #print(elves)
    #print(elvesset)
printpicture(elvesset)
empties(elvesset)
```
