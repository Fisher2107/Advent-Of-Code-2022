```python
#More efficient but still slow, gives correct answer however
with open('15.txt', 'r') as f:
    data = f.read().splitlines()

impossible=set()
sensors=set()
beacons=set()
for i in range(len(data)):
    x,y,equals = [],[],[]
    for j in range(len(data[i])):
        if data[i][j]==":":
            y.append(int(data[i][equals[-1]+1:j]))
        elif data[i][j]=="=":
            equals.append(j)
        elif data[i][j]==",":
            x.append(int(data[i][equals[-1]+1:j]))
        elif j == len(data[i])-1:
            y.append(int(data[i][equals[-1]+1:j+1]))
    data[i]=((x[0],y[0]),(x[1],y[1]))
    sensors.add(data[i][0])
    beacons.add(data[i][1])
    
def mod(x):
    if x<0:
        return -x
    else: return x
    
def add(x,y):
    return (x[0]+y[0],x[1]+y[1])
    
datadist=tuple(mod(data[i][0][0]-data[i][1][0])+mod(data[i][0][1]-data[i][1][1]) for i in range(len(data)))


def addimpossible(sensor,dist):
    right=add(sensor,(0,-dist))
    #list=[1,3,5,...,2*dist+1,2*dist-1,...,1]
    listt=[(i*2)+1 for i in range(dist+1)]+[(i*2)+1 for i in reversed(range(dist))]
    
    if right[1]==2000000: impossible.add(right)
    flip=False
    for i in range(2*dist):
    
        if not flip: right=add(right,(1,1))
        else: right=add(right,(-1,1))
        if right[1]==sensor[1]: flip =True
        if right[1]==2000000:
            for j in range(listt[i+1]):
                toadd=add(right,(-j,0))
                if toadd in sensors or toadd in impossible or toadd in beacons:
                    pass
                else:
                    impossible.add(toadd)

for i in range(len(data)):
    addimpossible(data[i][0],datadist[i])

count=len(impossible)
count
```

```python
#since we know only one exist we only have to search along the outline of the sensors detection radius
#for each sensor i outsiderange[i] will provide a set of all co-ordinates along its perimeter that are >=0 and <= 4 mil
outsiderange=set()

#new function will check for >=0 and <= 4 mil
def fits(x):
    if x[0]>=0 and x[0]<=4000001 and x[1]>=0 and x[1]<=4000001:
        return True
    return False

#Creating outside range list
for i in range(len(data)):
    sensor=data[i][0]
    dist=datadist[i]
    
    top=add(sensor,(0,-dist-1))
    right = add(top,(1,1))
    left = add(top,(-1,1))
    
    if fits(top): outsiderange.add(top)
    if fits(right): outsiderange.add(right)
    if fits(left): outsiderange.add(left)
    
    flip=False
    while True:
        if not flip: 
            right=add(right,(1,1))
            left=add(left,(-1,1))
        else: 
            right=add(right,(-1,1))
            left=add(left,(1,1))
            
        if right[1]==sensor[1]:  flip =True

        if fits(right): outsiderange.add(right)
        if fits(left): outsiderange.add(left)
        if left==right:
            break

for y in outsiderange:
    broken=False
    if y in beacons or y in impossible or y in sensors:
        continue
    for i in range(len(data)):
        distance=mod(data[i][0][0]-y[0])+mod(data[i][0][1]-y[1])
        if distance <= datadist[i]:
            broken = True
            break
    if not broken:
        print(y)
        print(y[0]*4000000+ y[1])
        break
```
