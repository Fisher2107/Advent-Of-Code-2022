```python
import re
with open('22.txt', 'r') as f:
    data=tuple(line for line in f.read().splitlines())
instructions=data[-1]
maxlen=0
for i in range((len(data)-1)):
    if len(data[i])>maxlen:
        maxlen=len(data[i])
print(maxlen)
maxdepth=len(data)-1
print(maxdepth)
#mapp will consist of cordinated of map followed with a number ex: (x,y):n n=1 implies dot and n=2 implies wall
mapp={}
for i in range((len(data)-1)):
    for j in range(len(data[i])):
        if data[i][j]==".":
            mapp[(j,i)]=1
        elif data[i][j]=="#":
            mapp[(j,i)]=2

intnum=tuple(int(i) for i in re.findall("[0-9-]+",instructions))
intletter=tuple(re.findall("[A-Z-]+",instructions))
Facing=("up","right","down","left")
facingnow=1
print(instructions[:100])
for i in range(len(data[0])):
    if (i,0) in mapp:
        location=(i,0)
        break
     
def turn(letter):
    global facingnow
    if letter=="R":
        facingnow+=1
        if facingnow>3:
            facingnow=0
        return
    if letter=="L":
        facingnow-=1
        if facingnow<0:
            facingnow=3
        return 

def move(num):
    global location
    def move1(x,y):
        if facingnow==0:
            return x,y-1
        elif facingnow==1:
            return x+1,y
        elif facingnow==2:
            return x,y+1
        elif facingnow==3:
            return x-1,y
            
    def loop(x,y):
        #print(x,y)
        if facingnow==0:
            return x,maxdepth-1
        elif facingnow==1:
            return 0,y
        elif facingnow==2:
            return x,0
        elif facingnow==3:
            return maxlen-1,y
        
    x,y = location[0],location[1]
    for i in range(num):
        location = (x,y)
        #print(1,location)
        x,y = move1(x,y)
        if (x,y) in mapp:
            if mapp[(x,y)]==2:
                return
        else:
            x,y = loop(x,y)
            while (x,y) not in mapp:
                x,y = move1(x,y)
            if mapp[(x,y)]==2:
                return 
    location = (x,y)
        
    
for i in range(len(intnum)-1):
    move(intnum[i])
    turn(intletter[i])
    #print(2,location)
    #print(3,facingnow)
move(intnum[-1])

print(Facing[facingnow])
answer=1000*(location[1]+1)+4*((location[0]+1))
print(answer+2)
```
